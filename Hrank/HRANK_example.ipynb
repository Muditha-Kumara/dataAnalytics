```python
import urllib
from bs4 import BeautifulSoup
from bs4.element import Comment

URL= "https://www.wiley.com/en-ie"

def Web_Scrapper_HTML(URL):
    request = urllib.request.Request(URL, headers = {"User-Agent": "Magic Browser"})
    con = urllib.request.urlopen(request)
    html = con.read().decode("utf-8")
    return html

def Web_scrapper_BeautifulSoup(URL):
    
    Raw_HTML = Web_Scrapper_HTML(URL)
    soup_html = BeautifulSoup(Raw_HTML, "lxml")
    text = soup_html.findAll(text = True)
    string = soup_html.findAll(string = True)
    tag_text = soup_html.findAll(name =True)
    visible_text = []
    tag_names = ['html','style', 'script', 'head', '[document]', 'img']
    
    for element in text:
        if element.parent.name not in tag_names:
            if not isinstance(element, Comment):
                visible_text.append(element.strip())
                
    Text = u" ".join(visible_text)
    text_lines = (line.strip() for line in Text.splitlines())
    chunks = (phrase.strip() for text_line in text_lines for phrase in text_line.split(" "))
    
    return u" ".join(chunk for chunk in chunks if chunk)

HTML_Text = Web_scrapper_BeautifulSoup(URL)
print(HTML_Text)
```

```python
import pickle

def Save_to_pickle(file_path, data):
    with open(file_path, "wb") as file:
        pickle.dump(data, file)
    print(f"Data successfully saved to {file_path}")


def Load_from_pickle(file_path):
    
    with open(file_path, "rb") as file:
        data = pickle.load(file)
    print(f"Data successfully loaded from {file_path}")
    return data

file_path = "PKL_Files/html_text.pkl"
Save_to_pickle(file_path, HTML_Text)
HTML_text = Load_from_pickle(file_path)
```

```python
import re
from nltk.corpus import stopwords

Special_Char_List =  re.compile(r"`|~|!|@|\#|\$|\%|\^|\&|\*|\(|\)|\-|\_|\=|\+|\[|\]|\{|\}|\\|\||\;|\:|\'|\"|\,|\<|\.|\>|\?|\/")

eng_stopword_list = set (stopwords.words("english"))
common_nouns = "gift debt free est dec big than who".split(" ")

def preprocess_text(HTML_text, common_nouns, Special_Char_List, eng_stopword_list):
    tokens_list = []   
    for token in HTML_text.split():
        token = token.lower().replace("'","")
        token = Special_Char_List.sub("", token.strip())
        
        if (
            len(token) > 1 and
            token not in eng_stopword_list and
            token not in common_nouns and
            not token.isdigit() and
            not token[0].isdigit()
        ):
            tokens_list.append(token)
       
    return tokens_list
    
tokens_list = preprocess_text(HTML_text,common_nouns,Special_Char_List,eng_stopword_list)
file_path = "PKL_Files/candidate_words.pkl"
Save_to_pickle(file_path, tokens_list)
candidate_words = Load_from_pickle(file_path)
candidate_words
```

```python
from textblob import TextBlob
noun_tags = ["NN", "NNS", "NNP", "NNPS"]
adj_tags = ["JJ", "JJR", "JJS"]
verb_tags = ["VB","VBD", "VBG","VBN","VBP","VBZ"]

def POS_Separator(candidate_words):
    nouns, adj, verb = [],[],[]
    for word in candidate_words:
        blob = TextBlob(word)
        tagged = blob.tags
        for word, tag in tagged:
            if tag in noun_tags:
                nouns.append(word)
            elif tag in adj_tags:
                adj.append(word)
            elif tag in verb_tags:
                verb.append(word)
    return nouns, adj, verb

nouns, adj, verb = POS_Separator(candidate_words)

print("Nouns:", nouns[:40])
print("Adjectives:", adj[:40])
print("Verbs:", verb[:40])
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def word_lemmatization(pos_tokens):
    lemmatized_words = []

    for word in pos_tokens:
        lemma = lemmatizer.lemmatize(word)
        lemmatized_words.append(lemma)

    return lemmatized_words


lemma_nouns = word_lemmatization(nouns)
lemma_adj   = word_lemmatization(adj)
lemma_verb  = word_lemmatization(verb)
```

```python
def clean_keywords_tokens(words):
    site_stop = {
        "guardian","news","world","search","hide","show","menu","view","home","back",
        "edition","international","latest","today","paper","jobs","subscriptions",
        "europe","uk","us","americas","asia","australia","africa","middle","east",
        "facebook","instagram","linkedin","tiktok","youtube",
        "s","dcr"
    }

    cleaned = []
    for w in words:
        w = w.lower()
        if len(w) < 3:
            continue
        if not w.isalpha():
            continue
        if w in site_stop:
            continue
        cleaned.append(w)
    return cleaned
```

```python
def Get_Synsets(candidate_words):
    words_with_synsets =[]
    words_without_synsets =[]

    for word in candidate_words:
        word_synonym = wn.synsets(word)
        if len(word_synonym)> 0:
            words_with_synsets.append(word)
        else:
            words_without_synsets.append(word)
    return words_with_synsets, words_without_synsets
        
words_with_synsets, words_without_synsets = Get_Synsets(lemma_nouns)
words_with_synsets
```

```python
import nltk
from sklearn.cluster import AgglomerativeClustering
from nltk.corpus import wordnet as wn

def compute_similarity_matrix(words):
    similarity_matrix = []

    for word1 in words:
        synsets1 = wn.synsets(word1)
        if not synsets1:
            similarity_matrix.append([1.0] * len(words))
            continue

        row = []

        for word2 in words:
            synsets2 = wn.synsets(word2)
            if not synsets2:
                row.append(1.0)
                continue
            
            similarity = synsets1[0].wup_similarity(synsets2[0])
            row.append(1.0 - (similarity if similarity is not None else 0.0))

        similarity_matrix.append(row)
    return similarity_matrix


def compute_clustering(sim_mat, num_clusters):
    clustering_model = AgglomerativeClustering(n_clusters=num_clusters,metric="precomputed",linkage="complete")
    return clustering_model.fit_predict(sim_mat)
```

```python
from collections import Counter


def extract_keywords(cluster_labels, words, words_fr, num_clusters):
    clusters = {i: [] for i in range(num_clusters)}

    for i, label in enumerate(cluster_labels):
        clusters[label].append((words[i], words_fr[words[i]]))

    keywords = []
    max_fr = max(words_fr.values())

    for cluster in clusters.values():
        cluster.sort(key=lambda x: x[1], reverse=True)
        keywords.append(cluster[0][0])

        for word, fr in cluster[1:]:
            if fr > 3 and fr > 0.2 * max_fr:
                keywords.append(word)

    return keywords

def get_clusters(words_fr, words, num_clusters=8):
    similarity_matrix = compute_similarity_matrix(words)
    cluster_labels = compute_clustering(similarity_matrix, num_clusters)
    Keywords = extract_keywords(cluster_labels, words, words_fr, num_clusters)
    return Keywords


from collections import defaultdict, Counter
clean_nouns = clean_keywords_tokens(lemma_nouns)

noun_counter = Counter(clean_nouns)
noun_counter_dict = dict(noun_counter)

TOP_N = 150
nouns = [w for w, fr in noun_counter.most_common(TOP_N)]

keywords = get_clusters(noun_counter_dict, nouns, num_clusters=8)
print("Cluster Keywords:", keywords)

def fr_adj_ver(lemma_words, N):
    count_lemmas = Counter(lemma_words)
    words = []
    for word, fr in count_lemmas.most_common(N):
        words.append(word)
    return words

fr_adj = fr_adj_ver(lemma_adj, 2)
fr_ver = fr_adj_ver(lemma_verb, 1)
Final_keywords = keywords + fr_adj + fr_ver
print(Final_keywords)
```
